```python
import sys
sys.path.insert(0, "./shared")

import pyodbc
import configparser as cfg
import pandas as pd
from datetime import datetime
import shared.athail as ab
import numpy as np
```

```python
# global Config

comparefrom_directory = "T:\Faith\compare\compare_from\\"
compareto_directory = "T:\Faith\compare\compare_to\\"
origfile_directory = "T:\Faith\compare\origails\\"

fromseedvalue = "880953"
toseedvalue = "902457"
#ffile = "AEL_M3Estimated_all_Choice_2021-12-31_"
#tfile = "AEL_M3Estimated_all_Choice_PGAAP_2021-12-31_"
#ffile = "AEL_M3Estimated_all_Fixed_2021-12-31_"
#tfile = "AEL_M3Estimated_all_Fixed_PGAAP_2021-12-31_"
#ffile = "AEL_M3Estimated_all_Indexed_2021-12-31_"
#tfile = "AEL_M3Estimated_all_Indexed_PGAAP_2021-12-31_"
#ffile = "EGL_M3Estimated_all_Fixed_2021-12-31_"
#tfile = "EGL_M3Estimated_all_Fixed_PGAAP_2021-12-31_"
ffile = "EGL_M3Estimated_all_Indexed_2021-12-31_"
#tfile = "EGL_M3Estimated_all_Indexed_PGAAP_2021-12-31_"

#ofile = "AEL_M3Estimated_all_Choice_2021-12-31_"
#ofile = "AEL_M3Estimated_all_Fixed_2021-12-31_"
#ofile = "AEL_M3Estimated_all_Indexed_2021-12-31_"
#ofile = "EGL_M3Estimated_all_Fixed_2021-12-31_"
ofile = "EGL_M3Estimated_all_Indexed_2021-12-31_"

comparefromfile = comparefrom_directory + ffile + fromseedvalue + "_PGAAP.ail2"
#comparetofile = compareto_directory + tfile + toseedvalue + ".ail2"
comparetofile = origfile_directory + ofile + fromseedvalue + ".ail2"

outputstring = "Comparing: \n" + comparefromfile + "\n" + comparetofile
ab.output_to_file(outputstring)
```

```python
#  comparison of values does not need datatypes.   Read_csv without dtypes used for expediency
from_ail            = pd.read_csv(  comparefromfile,
                                    sep     = '\t')

to_ail              = pd.read_csv(  comparetofile,
                                    sep     = '\t')

start = datetime.now()
starttime = str(start)

from_ail_total = len(from_ail)
to_ail_total = len(to_ail)

outputstring = "Total from Policies:  " + str(from_ail_total)
ab.output_to_file(outputstring)

outputstring = "Total to Policies: " + str(to_ail_total)
ab.output_to_file(outputstring)
```

```python

equalcheck = from_ail.equals(to_ail)
print("Dataframes are the same: ", equalcheck)

if equalcheck =='True':
    outputstring = "No Differences Identified"
    ab.output_to_file(outputstring)
else:
    outputstring = "Differences Found"
    ab.output_to_file(outputstring)
```

```python
# Identify the columns that have differences.
for column in from_ail:
    columnname = column
 
    if from_ail[column].equals(to_ail[column]):
        print(columnname + " match")
    else:
        outputstring = columnname + " Differ"
        ab.output_to_file(outputstring)
        print(outputstring)

```

```python

end = datetime.now()
endtime = "Total Time: " + str(end - start) + '\n\n'
ab.output_to_file(endtime)
print("Total Time: ", endtime)


```
